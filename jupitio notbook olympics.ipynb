```python
# see the fetails of the athelets who have won gold medals

goldmedals=atheletes_df[(atheletes_df.Medal=="Gold")]
goldmedals.head(10)

```

```python
# take values thet are different from nan 
goldmedals=goldmedals[np.isfinite(goldmedals['Age'])] 
```

```python
# athelets who won gold medal beyond the age of 60 years
goldmedals['ID'][goldmedals['Age']>60].count()
```

```python
# check for which these 6 gold medals have come from the age of 60 above catagory
sportingevent = goldmedals['Sport'][goldmedals['Age']>60]
sportingevent
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sportingevent)
plt.title("Gold Medals for Atheletics Over 60 years")
```

```python
# total gold medals from each countries 
goldmedals.Region.value_counts().reset_index(name="Medal").head(30)
```

```python
# create a plot to visualize the table
totalgoldmedals=goldmedals.Region.value_counts().reset_index(name="Medal").head(6)
g=sns.catplot(x="index",y="Medal",data=totalgoldmedals,height=5,kind="bar",palette="rocket")
g.despine(left=True)
g.set_xlabels("top 5 countries")
g.set_ylabels("Number of medals")
plt.title("gold metal per country")
```
